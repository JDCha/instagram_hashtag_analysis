```python
file_name = "#" + exercise_name +"_sum.txt"
file_1 = "/Users/noopy/FitCuration/" + file_name
text = open(file_1, 'r',-1,"UTF-8",errors="ignore").read()
text
```

```python
# get frequency analysis for each exercise
import emoji

def extract_emojis(str):
  emoji_string = ''.join(c for c in str if c in emoji.UNICODE_EMOJI)
  return emoji_string

emoji_string = extract_emojis(text)

from collections import Counter
common_emoji = Counter(emoji_string)
top_100_emojis = common_emoji.most_common(100)
top_100_emojis
```

```python
# -*- coding: utf-8 -*-
import re
def preprocessing(text):
    # 개행문자 제거
    hangul = re.compile('[^ ㄱ-ㅣ가-힣]+')
    result = hangul.sub('', text)
    return result
clean_text = preprocessing(text)
clean_text
```

```python
from konlpy.tag import Mecab
mecab = Mecab()
test_mecab = mecab.nouns(clean_text)
```

```python
stopwords_mecab = ['수','퀄리티','도시','분','전문','스타','년','원'] 

mecab_noun_list = []

for n in test_mecab:
    if n not in stopwords_mecab:
        mecab_noun_list.append(n)

from collections import Counter
mecab_nouns = Counter(mecab_noun_list)
# counted_nouns

tags = mecab_nouns.most_common(100)
tags        
```
